### 12번. 우유가 넘어지면? [17363]

https://intrepidgeeks.com/tutorial/if-bai-jun-17363-milk-falls-down

```python
N, M = map(int, input().split())
data = [list(map(str, input())) for _ in range(N)]
result = []
conv = {
    '.': '.', 'O': 'O', '-': '|', '|': '-', '/': '\\',
    '\\': '/', '^': '<', '<': 'v', 'v': '>', '>': '^'
}

for col in zip(*data):
    tmp = [conv[character] for character in col]
    result.append(tmp)

for row in reversed(result):
    print(*row, sep='')
```

```python
N=int(input()) 
S=input() # 처음 꺼
cnt=0
for i in range(N-1):
    SS=list(S)
    s=list(input()) # 나머지
    for i in range(len(S)):
        if S[i] in s:
            SS.remove(S[i])
            s.remove(S[i])
    if max(len(s),len(SS)) <=1:
        cnt+=1
print(cnt)
```

S는 처음 입력이고 SS는 S의 리스트, s는 밑에 값들임. 예제로 보면 GOD를 의미하는데 만약 S[0]인 D가 s안에 존재한다면 SS에서 D를 제거하고, s에서도 D를 제거한다. 이렇게 반복하다보면 마지막에 max 값을 구해준다. 이게 1 보다 작아야 같은 구성이 되게 되는거니까 그렇다면 +1을 해주는 것이다.
